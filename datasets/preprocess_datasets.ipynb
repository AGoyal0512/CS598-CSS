```python
import pandas as pd

train_df = pd.read_csv("../datasets/liar/train-extended.csv", header=0, names=['file', 'label', 'statement', 'subject', 'speaker', 'job_title', 'state_info', 'party_affiliation', 'barely_true', 'false', 'half_true', 'mostly_true', 'pants_fire', 'context'])
```

```python
train_df = train_df[train_df['label'].isin(['pants-fire', 'false', 'true'])]
train_df['label'] = train_df['label'].map({'false': False, 'true': True, 'pants-fire': False})
train_df.reset_index(drop=True, inplace=True)
```

```python
train_df
```

```python
test_df = pd.read_csv("../datasets/liar/test-extended.csv", header=0, names=['file', 'label', 'statement', 'subject', 'speaker', 'job_title', 'state_info', 'party_affiliation', 'barely_true', 'false', 'half_true', 'mostly_true', 'pants_fire', 'context'])
test_df = test_df[test_df['label'].isin(['pants-fire', 'false', 'true'])]
test_df['label'] = test_df['label'].map({'false': False, 'true': True, 'pants-fire': False})
test_df.reset_index(drop=True, inplace=True)
test_df
```

```python
# concatenate train and test
df = pd.concat([train_df, test_df], ignore_index=True)
df.to_csv("../datasets/liar/liar.csv", index=False)
```
